```python
from pathlib import Path
import pandas as pd


month = 1
year = 2024
path = Path("..") / "data" / "raw" / f"rides_{year}_{month:02}.csv"


df = pd.read_csv(path)
df.sort_values(by="started_at", inplace=True)


df.head()
```

```python

rides_cp['started_at'] = pd.to_datetime(rides_cp['started_at'])
rides_cp['ended_at'] = pd.to_datetime(rides_cp['ended_at'])


rides_cp['duration'] = (rides_cp['ended_at'] - rides_cp['started_at'])


duration_description = rides_cp['duration'].describe()
duration_description
```

```python
final_filter = duration_filter & filter_date_range
numbers_dropped = final_filter.shape[0] - sum(final_filter) 
numbers_dropped
numbers_dropped/final_filter.shape[0] * 100
```

```python
import pandas as pd
from pathlib import Path


raw_path = Path("..") / "data" / "raw"

# List to collect all monthly DataFrames
all_rides = []

# Iterate through all relevant CSV files
for csv_file in sorted(raw_path.glob("rides_2024_*.csv")):
    try:
        
        df = pd.read_csv(csv_file)
        # Parse datetime early for verification
        df["started_at"] = pd.to_datetime(df["started_at"], errors='coerce')
        all_rides.append(df)


    except Exception as e:
        print(f" Failed to process {csv_file.name}: {e}")

# Combine all months into a single DataFrame
df1 = pd.concat(all_rides, ignore_index=True)

```

```python
df1
```

```python
df1.fillna('',inplace=True)
invalid_ids = df1[df1['start_station_id'].str.contains(r'^\d+(\.\d+)?$', regex=True)]

# Display the filtered result
invalid_ids.head()
```

```python
df1['started_at'] = pd.to_datetime(df1['started_at'])  
january_records = df1[df1['started_at'].dt.month == 1]
january_records
```

```python
df1['started_at'] = pd.to_datetime(df1['started_at'])  
feb_records = df1[df1['started_at'].dt.month == 2]
feb_records
```

```python
df1['started_at'] = pd.to_datetime(df1['started_at'])  
Nov_records = df1[df1['started_at'].dt.month == 11]
Nov_records
```

```python
# 1. Count trips per (id, name) pair
station_counts = (
    df1
    .groupby(["start_station_id", "start_station_name"])
    .size()
    .reset_index(name="trip_count")
)

# 2. Sort descending and take top 3
top3 = station_counts.sort_values("trip_count", ascending=False).head(3)

# 3. Display just the id & name
print(top3[["start_station_id", "start_station_name"]])
```
